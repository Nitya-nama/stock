#### SOME VISUALISATION ANALYSIS

### 1.Plot Stock Price Trends


```python
plt.figure(figsize=(12, 6))
plt.plot(df.index, df["Close"], label="Closing Price", color="GREEN")
plt.title("Uber Stock Price Over Time")
plt.xlabel("Date")
plt.ylabel("Closing Price (USD)")
plt.legend()
plt.grid()
```

### 2.Moving Averages for Trend Analysis


```python

df["SMA_50"] = df["Close"].rolling(window=50).mean()  # 50-day SMA
df["SMA_200"] = df["Close"].rolling(window=200).mean()  # 200-day SMA

plt.figure(figsize=(12, 6))
plt.plot(df.index, df["Close"], label="Closing Price", alpha=0.6)
plt.plot(df.index, df["SMA_50"], label="50-day SMA", linestyle="dashed")
plt.plot(df.index, df["SMA_200"], label="200-day SMA", linestyle="dashed")
plt.title("Uber Stock Price with Moving Averages")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.legend()
plt.grid()
```

### 3. Trading Volume Analysis


```python

plt.figure(figsize=(12, 5))
plt.bar(df.index, df["Volume"], color="purple", alpha=0.5)
plt.title("Trading Volume Over Time")
plt.xlabel("Date")
plt.ylabel("Volume")
```

### 4.Daily Returns & Volatility Analysis



```python

df["Daily Returns"] = df["Close"].pct_change()

plt.figure(figsize=(10, 5))
sns.histplot(df["Daily Returns"].dropna(), bins=50, kde=True,color='black')
plt.title("Distribution of Daily Returns")
plt.xlabel("Daily Return")
plt.ylabel("Frequency")
```

### 5.Correlation Between Uber Stock & Trading Volume


```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df["Volume"], y=df["Close"],color='red')
plt.title("Uber Stock Price vs Trading Volume")
plt.xlabel("Trading Volume")
plt.ylabel("Closing Price")
```

### 6. Volatility (Rolling Standard Deviation of Returns)


```python

# Calculate the rolling volatility
df["Volatility"] = df["Daily Returns"].rolling(window=30).std()

plt.figure(figsize=(12, 6))
plt.plot(df.index, df["Volatility"], color="purple", label="30-Day Rolling Volatility")
plt.title("Uber Stock Volatility Over Time")
plt.xlabel("Date")
plt.ylabel("Volatility")
plt.legend()
plt.grid()
```

### 7. Relationship between different stock prices.



```python
df_subset = df[["Open", "High", "Low", "Close", "Volume"]]
sns.pairplot(df_subset, diag_kind="kde", corner=True)
plt.suptitle("Pairplot of Uber Stock Data", y=1.02)
```

### 8. Relationship Between Stock Price & Volume


```python
sns.jointplot(data=df, x="Volume", y="Close", kind="scatter", hue=None,color="orange")
plt.suptitle("Uber Stock Price vs Trading Volume", y=1.02)
```

### 9.High correlation between Open, High, Low, and Close prices.


```python
plt.figure(figsize=(8, 6))
sns.heatmap(df_subset.corr(), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap of Uber Stock Data")
```

### 10.Distribution of Daily Returns





```python
sns.kdeplot(df["Daily Returns"].dropna(), fill=True, color="darkgreen")
plt.title("Density Plot of Uber's Daily Returns")
plt.xlabel("Daily Return")
plt.ylabel("Density")
```
